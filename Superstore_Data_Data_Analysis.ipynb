```python
# Let's drop the Postal Code column 

dfc = dfc.drop(columns = ['Postal Code'], axis = 1)
```

```python
# Let's check the value counts of the columns which are having unique values less than 20 :- 

for i in [col for col in dfc.columns if dfc[col].nunique()<20]:
    print(f"{i.upper()} : -")
    print('---'*10)
    print(dfc[i].value_counts().reset_index())
    print('\n')
```

```python
#  let's get the 'Sales' ,'Profit' and 'Profit %' of different region :- 

dfc_region_SP = pd.pivot_table( data = dfc,
                                index = ['Region'],
                                values = ['Profit', 'Sales'],
                                aggfunc = 'sum').reset_index()

# let's get the profit percentage region wise

dfc_region_SP['Profit %'] = round((dfc_region_SP['Profit']/(dfc_region_SP['Sales']-dfc_region_SP['Profit']))*100,2)

# Let's print the dataframe

dfc_region_SP.style.background_gradient(subset = ['Profit','Sales','Profit %'], cmap = 'GnBu').format({
                                                                                 'Profit': '${0:,.2f}', 'Sales': '${0:,.2f}', 'Profit %': '{0:,.2f}%' })
```

```python
# Let's plot the Region wise Profit % 


plt.style.use('fivethirtyeight')
plt.bar(x = dfc_region_SP['Region'], height = dfc_region_SP['Profit %'], color = ['red','blue','green','purple'])
plt.title('Region VS Profit %', fontsize = 12, color = 'black')
plt.xlabel('Region', fontsize = 12, color = 'black')
plt.ylabel('Profit %', fontsize = 12, color = 'black')
for i,v in enumerate(dfc_region_SP['Profit %']):
    plt.text(i,v,"{}%".format(str(v)),ha='center',va='bottom',fontsize='medium',color = 'red', backgroundcolor = 'yellow')
plt.show()
```

**Observation:-**

* **West region is having most profit % & central region generates very less profit so maybe we have to reduce our sales operations in central region.**

```python
# Let's get the Profit % of different states : - 

dfc_state_region_SP = pd.pivot_table( data = dfc,
                                index = ['Region', 'State'],
                                values = ['Profit', 'Sales'],
                                aggfunc = 'sum').reset_index()

# let's get the profit percentage state wise

dfc_state_region_SP['Profit %'] = round((dfc_state_region_SP['Profit']/(dfc_state_region_SP['Sales']-dfc_state_region_SP['Profit']))*100,2)

# Let's see the Dataframe

dfc_state_region_SP.sort_values(by = 'Profit %', ascending = False).style.background_gradient(subset = 'Profit %', cmap = 'YlOrRd').format({
                                                                                 'Profit': '${0:,.2f}', 'Sales': '${0:,.2f}', 'Profit %': '{0:,.2f} %'})
```

**Observation:-**

* **As we can see from above dataframe that District of Columbia, Delaware, Minnesota, Maine, Indiana these are some of the states which are having a good profit % inspite of their less sales and profit.**

* **Maximum states which are having a good profit %, are from Central, East, south.**


* **So, maybe we need to focus more on these states which are getting us good profit %.**

```python
# Let's find out lowest performing states:-  

# Let's create a function to highlight profit, sales and profit %

def highlight_sales(value):
  if isinstance(value, str):
    return None
  if value<0:
    color = 'red'
  else:
    color = 'yellow'
  return f'color: {color}'

# Let's see the pivot table for region and state wise profit and sales values , profit less than 0 is marked in red color

dfc_state_region_SP.sort_values(by = 'Profit %').loc[dfc_state_region_SP['Profit %']<0].reset_index(drop=True).style.applymap(highlight_sales).format({'Profit': '${0:,.2f}', 
                                                                                                                                'Sales': '${0:,.2f}', 
                                                                                                                                'Profit %': '{0:,.2f} %'})
```

**Observation:-**

* **From above as we can see 10 worst performing states, which are having negative profit %.**

```python
# Let's find out maximum profit states region wise:-

s = pd.pivot_table( data = dfc,
                index = ['Region', 'State'],
                values = ['Profit'],
                aggfunc = 'sum',
                ).sort_values(by =['Region','Profit'], ascending = False)

highest_profit_state = s.reset_index().loc[s.reset_index().groupby(["Region"])["Profit"].idxmax()].sort_values(by = 'Profit',ascending = False)

# Let's highlight most profit states for the respective Region : - 
highest_profit_state.style.background_gradient(subset = 'Profit', cmap = 'YlOrRd').format({
                                                                                 'Profit': '${0:,.2f}'})
```

**Observation:-**

* **As we can see from above in West region California, in East region New York, in Central region Michigan and in South region Virginia is having the highest Profit amount.**


```python
# Let's check the cities according to every state which are having very good profit %, and profit and sales value are also high : -


k = pd.pivot_table( data = dfc,
                index = ['State','City'],
                values = ['Sales','Profit'],
                aggfunc = 'sum'
                ).reset_index()
# let's get the profit percentage state wise

k['Profit %'] = round((k['Profit']/(k['Sales']-k['Profit']))*100,2)

# Let's get the state wise cities which has most profit %
k = k.loc[k.groupby(["State"])["Profit %"].idxmax()].sort_values(by = 'Profit %', ascending = False)

def highlight_sales_profit(value):
  '''
  This function will highlight the profit and sales values which are above $500
  '''
  if isinstance(value, str):
    return None
  if value>500:
    color = 'yellow'
  else:
    color = 'red'
  return f'color: {color}'

# Let's see the dataframne for top cities with good profit % where profit and sales are greater than atleast $500

Top_city_profit_sales = k.loc[(k['Sales']>500) & (k['Profit']>500)].reset_index(drop=True).style.applymap(highlight_sales_profit).format({'Profit': '${0:,.2f}',
                                                                                                                                          'Sales': '${0:,.2f}',
                                                                                                                                          'Profit %': '{0:,.2f} %'})
# Let's check the dataframe 

Top_city_profit_sales
```

**Observation:-**

* **As we can see from above that in every state there is one top performing city is present, so above is the 14 top performing cities which gives us a good profit % and sales and profit amount are also good.**

```python
# Let's find out which sity in the state is having highest profit : -

d = pd.pivot_table( data = dfc,
                index = ['State','City'],
                values = ['Profit'],
                aggfunc = 'sum',
                ).sort_values(by =['State','Profit'], ascending = False)
                
highest_profit_City = d.reset_index().loc[d.reset_index().groupby(["State"])["Profit"].idxmax()].sort_values(by = 'Profit', ascending = False)

# Let's highlight most profit City for the respective State : - 
highest_profit_City.style.bar().format({'Profit': '${0:,.2f}'})
```

```python
# first let's create Month/Day/Year columns in the dataset

dfc['Month'] = dfc['Order Date'].dt.month
dfc['Day'] = dfc['Order Date'].dt.day
dfc['Year'] = dfc['Order Date'].dt.year
dfc['Month_name'] = dfc['Order Date'].dt.month_name()
```

```python
# Let's check the dataset 

dfc.head()
```

```python
# Let's check how the sales are going on the basis of different months:- 

# Sales comparion monthly basis
year_month_sale = pd.pivot_table(data = dfc,
               index = ['Year','Month_name'],
               values = ['Sales'],
               aggfunc = 'sum').unstack(0).reset_index()

# Let's sort the dataframe on the basis of month

month_dict = {'January':1,'February':2,'March':3, 'April':4, 'May':5, 'June':6, 'July':7, 'August':8, 'September':9, 'October':10, 'November':11, 'December':12}
year_month_sale.sort_values('Month_name', key = lambda x : x.apply (lambda x : month_dict[x]), inplace = True)

# Let's check the dataframe 
year_month_sale.style.background_gradient(subset = ['Sales'], cmap = 'YlOrRd').format({
                                                                                  ('Sales',2014): '${0:,.2f}',
                                                                                   ('Sales',2015): '${0:,.2f}',
                                                                                   ('Sales',2016): '${0:,.2f}', 
                                                                                   ('Sales',2017): '${0:,.2f}'})
```

```python
# Let's visualize month wise sales comparison on yearly basis

plt.style.use('dark_background')
fig,ax = plt.subplots(figsize=(12,4))

ax.plot(year_month_sale['Month_name'], year_month_sale[('Sales',2014)], linewidth=2, color='red', marker ='x', markersize=6, label= '2014')
ax.plot(year_month_sale['Month_name'], year_month_sale[('Sales',2015)], linewidth=2, color='lightgreen', marker ='o', markersize=6, label= '2015')
ax.plot(year_month_sale['Month_name'], year_month_sale[('Sales',2016)], linewidth=2, color='yellow', marker ='*', markersize=6, label= '2016')
ax.plot(year_month_sale['Month_name'], year_month_sale[('Sales',2017)], linewidth=2, color='white', marker ='+', markersize=6, label= '2017')

ax.set_title("Monthly Sales Per Year",fontsize=12)
ax.set_xlabel('Month', fontsize=12)
ax.set_ylabel('Sales', fontsize=12)
plt. grid(False)
plt.xticks(fontsize = 10)

ax.legend(loc='upper left',fontsize=12)

plt.show()
```

**Observation :-**

* **Sales at the end of the months are high compare to initial months of a year, The reason maybe because of maximum festival seasons are in the end of year.**

* **2017 sales are higher compare to 2016/2015/2014.**
